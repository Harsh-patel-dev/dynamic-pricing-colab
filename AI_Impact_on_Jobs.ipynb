```python
# MERGE THE TWO COLUMNS WITH THE SAME NUMBER OF THE COLUMNS LIKE INT, STRING , FLOAT 

import pandas as pd
# Create DataFrame 1
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Harsh', 'Ran', 'Anna']
})

# Create DataFrame 2
df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Age': [12, 15, 17]
})

# Merge the DataFrames
result = pd.merge(df1, df2, on='ID')

print(result)
```

# CONCAT

```python
pd.concat([df1,df2],axis=0)
```
